```python
import yfinance as yf
import pandas as pd

class YahooFinanceDataLoader:
    def __init__(self):
        self.data = pd.DataFrame()  # To store the ticker info data
    
    def load_ticker_info(self, stock_symbols):
        # List to store processed data
        ticker_info_list = []

        for symbol in stock_symbols:
            try:
                # Get ticker info
                ticker = yf.Ticker(symbol)
                ticker_info = ticker.info

                # Remove 'companyOfficers' field if it exists
                ticker_info.pop('companyOfficers', None)

                # Append the remaining info to the list
                ticker_info_list.append(ticker_info)
                print(f"Successfully loaded data for {symbol}")
            except Exception as e:
                print(f"Error loading data for {symbol}: {e}")
        
        # Convert the list of dictionaries into a DataFrame
        self.data = pd.DataFrame(ticker_info_list)

    def run(self, stock_file):
        # Read stock symbols from the file
        with open(stock_file, 'r') as file:
            stock_symbols = [line.strip() for line in file if line.strip()]
        
        if not stock_symbols:
            print("No stock symbols found in the file.")
            return
        
        # Load ticker info
        self.load_ticker_info(stock_symbols)
        print("Data loading completed.")

if __name__ == "__main__":
    # Path to the stock symbols file
    stock_file = r"C:\Users\nisha\Documents\PythonProjects\TradingAnalysis\src\config\stocks.txt"

    # Create an instance of the YahooFinanceDataLoader and run it
    loader = YahooFinanceDataLoader()
    loader.run(stock_file)

    # Save the resulting data to a CSV file
    if not loader.data.empty:
        loader.data.to_csv("ticker_info.csv", index=False)
        print("Ticker info saved to ticker_info.csv")
```
